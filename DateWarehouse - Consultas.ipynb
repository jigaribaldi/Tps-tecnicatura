# **TIPO DE CONTRATO PARA UN CLIENTE EN UNA FECHA DETERMINADA**

-- 1. Búsqueda de tipo de contrato para un cliente en cierta fecha


```python
# Convertir a tipo fecha las columnas necesarias en contratos_df
contratos_df['FechaInicio'] = pd.to_datetime(contratos_df['FechaInicio'], errors='coerce')
contratos_df['FechaFin'] = pd.to_datetime(contratos_df['FechaFin'], errors='coerce')

# Parámetros de búsqueda (MODIFICAR SEGÚN SE REQUIERA PARA LA CONSULTA)
empresa_id = 4
mes = 2
anio = 2023

# Definir el rango de fechas
fecha_inicio = pd.to_datetime(f'{anio}-{mes}-01')
fecha_fin = fecha_inicio + pd.offsets.MonthEnd(1)

# Filtrar el DataFrame para encontrar el contrato correspondiente
resultado_contratos = contratos_df[
    (contratos_df['EmpresaID'] == empresa_id) &
    (contratos_df['FechaInicio'] <= fecha_fin) &
    ((contratos_df['FechaFin'].isna()) | (contratos_df['FechaFin'] >= fecha_inicio))
]

# Realizar un merge para incluir el nombre de la empresa
resultado_contratos = resultado_contratos.merge(empresas_df[['EmpresaId', 'EmpresaNombre']],
                                                  left_on='EmpresaID',
                                                  right_on='EmpresaId',
                                                  how='left')

# Mostrar el resultado incluyendo el nombre de la empresa y el ContratoID
print(resultado_contratos[['ContratoID', 'EmpresaID', 'EmpresaNombre', 'NombreContrato', 'FechaInicio', 'FechaFin', 'Activo', 'Version']])
```

# **Conteo de contratos de cada tipo en un mes y año**

-- 3. Conteo de contratos de cada tipo en un mes y año


```python
# Convertir a tipo fecha las columnas necesarias en contratos_df
contratos_df['FechaInicio'] = pd.to_datetime(contratos_df['FechaInicio'], errors='coerce')
contratos_df['FechaFin'] = pd.to_datetime(contratos_df['FechaFin'], errors='coerce')

# Parámetros de búsqueda (MODIFICAR SEGÚN SE REQUIERA PARA LA CONSULTA)
mes = 7
anio = 2023

# Definir el rango de fechas
fecha_inicio = pd.to_datetime(f'{anio}-{mes}-01')
fecha_fin = fecha_inicio + pd.offsets.MonthEnd(1)

# Filtrar el DataFrame para contar los contratos de cada tipo en el rango de fechas
conteo_contratos = contratos_df[
    (contratos_df['FechaInicio'] <= fecha_fin) &
    ((contratos_df['FechaFin'].isna()) | (contratos_df['FechaFin'] >= fecha_inicio))
].groupby(['TipoContrato', 'NombreContrato']).size().reset_index(name='ConteoContratos')

# Ordenar el resultado por TipoContrato y NombreContrato
conteo_contratos = conteo_contratos.sort_values(by=['TipoContrato', 'NombreContrato'])

# Mostrar el resultado final de la sumatoria total de contratos por tipo
print(conteo_contratos)
```

# **Consultas de facturación**

FACTURACION EN PESOS Y EN DOLARES PARA UNA EMPRESA EN UN TIEMPO DETERMINADO

```python
# Parámetros de búsqueda (MODIFICAR SEGÚN SE REQUIERA PARA LA CONSULTA)
# FECHA DE CONSULTA ('aaaa-mm-dd')
fecha_desde = '2019-01-01'
fecha_hasta = '2021-12-31'
empresa_id = 1

# Filtrar los tickets según la fecha de cierre y el ID de empresa
ticket_df['TicketFechaCierre'] = pd.to_datetime(ticket_df['TicketFechaCierre'])  # Asegurar que es de tipo datetime
tickets_filtrados = ticket_df[
    (ticket_df['TicketFechaCierre'] >= fecha_desde) &
    (ticket_df['TicketFechaCierre'] <= fecha_hasta) &
    (ticket_df['EmpresaId'] == empresa_id)
]

# Unir con la tabla de dólares para obtener la conversión a dólares
# Aseguramos que ambas fechas sean del mismo tipo para la unión
dolar_df['Fecha'] = pd.to_datetime(dolar_df['Fecha'])
tickets_dolar = tickets_filtrados.merge(
    dolar_df,
    left_on=tickets_filtrados['TicketFechaCierre'].dt.date,
    right_on=dolar_df['Fecha'].dt.date,
    how='left'
)

# Unir con la tabla de empresas para obtener el nombre de la empresa
tickets_completos = tickets_dolar.merge(
    empresas_df,
    left_on='EmpresaId',
    right_on='EmpresaId',
    how='left'
)

# Calcular los valores de facturación en pesos y en dólares
tickets_completos['TotalFacturadoPesos'] = tickets_completos['Valor']
tickets_completos['TotalFacturadoDolares'] = tickets_completos['Valor'] / tickets_completos['ValorDolar']

# Agrupar por nombre de empresa y calcular las métricas
resultado = tickets_completos.groupby('EmpresaNombre').agg(
    CantidadTickets=('TicketId', 'count'),
    TotalFacturadoPesos=('TotalFacturadoPesos', 'sum'),
    TotalFacturadoDolares=('TotalFacturadoDolares', 'sum')
).reset_index()

# Agregar las fechas de inicio y fin de la consulta
resultado['FechaDesde'] = fecha_desde
resultado['FechaHasta'] = fecha_hasta

# Mostrar el resultado
print(resultado)
```

# **Consulta que valida periodo del contrato**

```python
# Convertir a tipo fecha las columnas necesarias
ticket_df['TicketFechaApertura'] = pd.to_datetime(ticket_df['TicketFechaApertura'], errors='coerce')
ticket_df['TicketFechaCierre'] = pd.to_datetime(ticket_df['TicketFechaCierre'], errors='coerce')
contratos_df['FechaInicio'] = pd.to_datetime(contratos_df['FechaInicio'], errors='coerce')
contratos_df['FechaFin'] = pd.to_datetime(contratos_df['FechaFin'], errors='coerce')

# Filtrar tickets que están cerrados
tickets_filtrados = ticket_df[ticket_df['TicketFechaCierre'].notnull()]

# Realizar un merge entre tickets y contratos basado en la EmpresaId y las fechas
resultado = (
    tickets_filtrados
    .merge(contratos_df, left_on='EmpresaId', right_on='EmpresaID', how='inner', suffixes=('', '_contrato'))
    .loc[lambda df: (df['TicketFechaCierre'] >= df['FechaInicio']) &
                (df['TicketFechaCierre'] <= df['FechaFin'].fillna(pd.Timestamp('now')))]
)

# Realizar un merge para obtener el nombre de la empresa
resultado = resultado.merge(empresas_df[['EmpresaId', 'EmpresaNombre']], left_on='EmpresaId', right_on='EmpresaId', how='inner')

# Seleccionar las columnas finales deseadas, usando la columna 'ContratoID' correcta
resultado_final = resultado[['TicketId', 'TicketFechaApertura', 'TicketFechaCierre', 'ContratoID', 'EmpresaNombre', 'FechaInicio', 'FechaFin']]

# Mostrar el resultado en un DataFrame
df_resultado = pd.DataFrame(resultado_final)

# Mostrar el DataFrame
df_resultado
```
